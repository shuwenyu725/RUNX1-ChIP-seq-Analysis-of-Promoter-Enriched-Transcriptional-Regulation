```python
%%bash
set -euxo pipefail

mkdir -p results/qc

S1="results/bigwig/IP_rep1.bw"
S2="results/bigwig/IP_rep2.bw"

test -f "$S1" && test -f "$S2"

computeMatrix scale-regions \
  -S "$S1" "$S2" \
  -R refs/hg38_genes.bed \
  --beforeRegionStartLength 2000 \
  --regionBodyLength 5000 \
  --afterRegionStartLength 2000 \
  --skipZeros --missingDataAsZero \
  -o results/qc/allgenes_matrix.gz \
  --verbose
```

```python
%%bash
plotProfile \
  -m results/qc/allgenes_matrix.gz \
  -out results/qc/allgenes_profile.png \
  --perGroup --legendLocation upper-right \
  --plotTitle "IP coverage over genes (hg38)"
```
